```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Product type', y='Price', data=df)
plt.title('Distribuição de Preços por Tipo de Produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Preço')
plt.show()
```

```python
media = df.groupby('Product type')['Price'].mean()
print(media)
```

```python
grouped_df = df.groupby('Product type')['Revenue generated'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Product type', y='Revenue generated', data=grouped_df, palette='Greys')

plt.title('Receita gerada por tipo de produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Receita gerada')
```

```python
# Gráfico de barras mostrando o total de produtos vendidos ("Number of products sold") para cada tipo de produto.
# Number of products sold
# Product type

grouped_df = df.groupby('Product type')['Number of products sold'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Product type', y='Number of products sold', data=grouped_df, palette='Reds')

plt.title('Número de produtos vendidos por tipo de produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Número de produtos vendidos')
```

```python
# Gráfico de barras mostrando a disponibilidade média 
# ("Availability") de produtos para cada tipo de produto.
# Availability
grouped_df_availability = df.groupby('Product type')['Availability'].mean().reset_index()
plt.figure(figsize=(10, 6))

sns.barplot(x='Product type', y='Availability', data=grouped_df_availability, palette='Oranges')
plt.title('Disponibilidade média por tipo de produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Disponibilidade média')
```

```python
# FORMA ALTERNATIVA:
plt.figure(figsize=(12, 8))
sns.barplot(x='Product type', y='Availability', data=df)
plt.title('Disponibilidade média por tipo de produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Disponibilidade média')
```

```python
# Análise de Demografia dos Clientes:
# Um gráfico de pizza mostrando a distribuição demográfica dos clientes ("Customer demographics").
# Customer demographics
Dados_demográficos_do_cliente = df['Customer demographics'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(Dados_demográficos_do_cliente, labels=Dados_demográficos_do_cliente.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição Demográfica dos Clientes')
```

```python
# Análise de Demografia dos Clientes:
# Um gráfico de pizza mostrando a distribuição demográfica dos clientes ("Customer demographics").
# Customer demographics
Dados_demográficos_do_cliente = df['Customer demographics'].value_counts().drop('Unknown', errors='ignore')

plt.figure(figsize=(8, 8))
plt.pie(Dados_demográficos_do_cliente, labels=Dados_demográficos_do_cliente.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição Demográfica dos Clientes')
```

```python
# Níveis de Estoque por Tipo de Produto:
# Um gráfico de barras para mostrar os níveis de estoque 
# ("Stock levels") para cada tipo de produto.
# O que precisa?  Product type & Stock levels - Níveis de estoque

grouped_df = df.groupby('Product type')['Stock levels'].sum().reset_index()
plt.figure(figsize=(10, 8))
sns.barplot(x='Product type', y='Stock levels', data=grouped_df, palette='Purples')

plt.title('Níveis de Estoque por Tipo de Produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Níveis de Estoque')
```

```python
# Relação entre Tempo de Espera e Quantidade de Pedidos:
# Um gráfico de dispersão (scatter plot) para mostrar a relação 
# entre os tempos de espera ("Lead times") e as quantidades de pedidos 
# ("Order quantities").
# Lead times - Prazo de entrega
# Order quantities - Quantidade de pedidos

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lead times', y='Order quantities', data=df, alpha=1)

plt.title('Relação entre Lead Times e Quantidades de Pedidos')
plt.xlabel('Prazos de entrega (dias)')
plt.ylabel('Quantidades do pedido')
```

```python
# Custos de Envio por Transportadora:
# custos médios de envio ("Shipping costs") para cada transportadora ("Shipping carriers").
# Shipping costs - Custos de envio
# Shipping carriers - Transportadoras


grouped_df_shipping = df.groupby('Shipping carriers')['Shipping costs'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='Shipping carriers', y='Shipping costs', data=grouped_df_shipping, palette='viridis')

plt.title('Custos médios de envio por transportadora')
plt.xlabel('Transportadoras')
plt.ylabel('Custos médios de envio')
```

```python
# Custos de Envio por Transportadora:
# custos médios de envio ("Shipping costs") para cada transportadora ("Shipping carriers").
# Shipping costs - Custos de envio
# Shipping carriers - Transportadoras


grouped_df_shipping = df.groupby('Shipping carriers')['Shipping costs'].sum().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='Shipping carriers', y='Shipping costs', data=grouped_df_shipping, palette='viridis')

plt.title('Custos médios de envio por transportadora')
plt.xlabel('Transportadoras')
plt.ylabel('Custos médios de envio')
```

```python
df.loc[:,['Supplier name','Production volumes']]
```

```python
# Tempos de Produção por Fornecedor:
# tempos de produção ("Production volumes" e "Manufacturing lead time") 
# para cada fornecedor ("Supplier name").
# Supplier name - Nome do fornecedor
# Production volumes - Volume de produção
# Manufacturing lead time - Tempo de fabricação


grupo_df = df.groupby('Supplier name')['Production volumes'].sum().reset_index()
plt.figure(figsize=(10, 8))
sns.barplot(x='Supplier name', y='Production volumes', data=grupo_df, palette='Purples')

plt.title('Nome do fornecedor por Volume de produção')
plt.xlabel('Nome do fornecedor')
plt.ylabel('Volume de produção')
```

```python
# Tempos de Produção por Fornecedor:
# tempos de produção ("Production volumes" e "Manufacturing lead time") 
# para cada fornecedor ("Supplier name").
# Supplier name - Nome do fornecedor
# Production volumes - Volume de produção
# Manufacturing lead time - Tempo de fabricação


grupo_df = df.groupby('Supplier name')['Manufacturing lead time'].sum().reset_index()
plt.figure(figsize=(10, 8))
sns.barplot(x='Supplier name', y='Manufacturing lead time', data=grupo_df, palette='Purples')

plt.title('Nome do fornecedor por Tempo de fabricação')
plt.xlabel('Nome do fornecedor')
plt.ylabel('Tempo de fabricação')
```

```python
# Number of products sold by Location

grouped_df = df.groupby('Location')['Number of products sold'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Location', y='Number of products sold', data=grouped_df, palette='Greens')

```
