```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Product type', y='Price', data=df)
plt.title('DistribuiÃ§Ã£o de PreÃ§os por Tipo de Produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('PreÃ§o')
plt.show()
```

```python
media = df.groupby('Product type')['Price'].mean()
print(media)
```

```python
grouped_df = df.groupby('Product type')['Revenue generated'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Product type', y='Revenue generated', data=grouped_df, palette='Greys')

plt.title('Receita gerada por tipo de produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Receita gerada')
```

```python
# GrÃ¡fico de barras mostrando o total de produtos vendidos ("Number of products sold") para cada tipo de produto.
# Number of products sold
# Product type

grouped_df = df.groupby('Product type')['Number of products sold'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Product type', y='Number of products sold', data=grouped_df, palette='Reds')

plt.title('NÃºmero de produtos vendidos por tipo de produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('NÃºmero de produtos vendidos')
```

```python
# GrÃ¡fico de barras mostrando a disponibilidade mÃ©dia 
# ("Availability") de produtos para cada tipo de produto.
# Availability
grouped_df_availability = df.groupby('Product type')['Availability'].mean().reset_index()
plt.figure(figsize=(10, 6))

sns.barplot(x='Product type', y='Availability', data=grouped_df_availability, palette='Oranges')
plt.title('Disponibilidade mÃ©dia por tipo de produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Disponibilidade mÃ©dia')
```

```python
# FORMA ALTERNATIVA:
plt.figure(figsize=(12, 8))
sns.barplot(x='Product type', y='Availability', data=df)
plt.title('Disponibilidade mÃ©dia por tipo de produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('Disponibilidade mÃ©dia')
```

```python
# AnÃ¡lise de Demografia dos Clientes:
# Um grÃ¡fico de pizza mostrando a distribuiÃ§Ã£o demogrÃ¡fica dos clientes ("Customer demographics").
# Customer demographics
Dados_demogrÃ¡ficos_do_cliente = df['Customer demographics'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(Dados_demogrÃ¡ficos_do_cliente, labels=Dados_demogrÃ¡ficos_do_cliente.index, autopct='%1.1f%%', startangle=140)
plt.title('DistribuiÃ§Ã£o DemogrÃ¡fica dos Clientes')
```

```python
# AnÃ¡lise de Demografia dos Clientes:
# Um grÃ¡fico de pizza mostrando a distribuiÃ§Ã£o demogrÃ¡fica dos clientes ("Customer demographics").
# Customer demographics
Dados_demogrÃ¡ficos_do_cliente = df['Customer demographics'].value_counts().drop('Unknown', errors='ignore')

plt.figure(figsize=(8, 8))
plt.pie(Dados_demogrÃ¡ficos_do_cliente, labels=Dados_demogrÃ¡ficos_do_cliente.index, autopct='%1.1f%%', startangle=140)
plt.title('DistribuiÃ§Ã£o DemogrÃ¡fica dos Clientes')
```

```python
# NÃ­veis de Estoque por Tipo de Produto:
# Um grÃ¡fico de barras para mostrar os nÃ­veis de estoque 
# ("Stock levels") para cada tipo de produto.
# O que precisa?  Product type & Stock levels - NÃ­veis de estoque

grouped_df = df.groupby('Product type')['Stock levels'].sum().reset_index()
plt.figure(figsize=(10, 8))
sns.barplot(x='Product type', y='Stock levels', data=grouped_df, palette='Purples')

plt.title('NÃ­veis de Estoque por Tipo de Produto')
plt.xlabel('Tipo de Produto')
plt.ylabel('NÃ­veis de Estoque')
```

```python
# RelaÃ§Ã£o entre Tempo de Espera e Quantidade de Pedidos:
# Um grÃ¡fico de dispersÃ£o (scatter plot) para mostrar a relaÃ§Ã£o 
# entre os tempos de espera ("Lead times") e as quantidades de pedidos 
# ("Order quantities").
# Lead times - Prazo de entrega
# Order quantities - Quantidade de pedidos

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lead times', y='Order quantities', data=df, alpha=1)

plt.title('RelaÃ§Ã£o entre Lead Times e Quantidades de Pedidos')
plt.xlabel('Prazos de entrega (dias)')
plt.ylabel('Quantidades do pedido')
```

```python
# Custos de Envio por Transportadora:
# custos mÃ©dios de envio ("Shipping costs") para cada transportadora ("Shipping carriers").
# Shipping costs - Custos de envio
# Shipping carriers - Transportadoras


grouped_df_shipping = df.groupby('Shipping carriers')['Shipping costs'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='Shipping carriers', y='Shipping costs', data=grouped_df_shipping, palette='viridis')

plt.title('Custos mÃ©dios de envio por transportadora')
plt.xlabel('Transportadoras')
plt.ylabel('Custos mÃ©dios de envio')
```

```python
# Custos de Envio por Transportadora:
# custos mÃ©dios de envio ("Shipping costs") para cada transportadora ("Shipping carriers").
# Shipping costs - Custos de envio
# Shipping carriers - Transportadoras


grouped_df_shipping = df.groupby('Shipping carriers')['Shipping costs'].sum().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='Shipping carriers', y='Shipping costs', data=grouped_df_shipping, palette='viridis')

plt.title('Custos mÃ©dios de envio por transportadora')
plt.xlabel('Transportadoras')
plt.ylabel('Custos mÃ©dios de envio')
```

```python
# Tempos de ProduÃ§Ã£o por Fornecedor:
# tempos de produÃ§Ã£o ("Production volumes" e "Manufacturing lead time") 
# para cada fornecedor ("Supplier name").
# Supplier name - Nome do fornecedor
# Production volumes - Volume de produÃ§Ã£o
# Manufacturing lead time - Tempo de fabricaÃ§Ã£o


grupo_df = df.groupby('Supplier name')['Production volumes'].sum().reset_index()
plt.figure(figsize=(10, 8))
sns.barplot(x='Supplier name', y='Production volumes', data=grupo_df, palette='Purples')

plt.title('Nome do fornecedor por Volume de produÃ§Ã£o')
plt.xlabel('Nome do fornecedor')
plt.ylabel('Volume de produÃ§Ã£o')
```

```python
# Tempos de ProduÃ§Ã£o por Fornecedor:
# tempos de produÃ§Ã£o ("Production volumes" e "Manufacturing lead time") 
# para cada fornecedor ("Supplier name").
# Supplier name - Nome do fornecedor
# Production volumes - Volume de produÃ§Ã£o
# Manufacturing lead time - Tempo de fabricaÃ§Ã£o


grupo_df = df.groupby('Supplier name')['Manufacturing lead time'].sum().reset_index()
plt.figure(figsize=(10, 8))
sns.barplot(x='Supplier name', y='Manufacturing lead time', data=grupo_df, palette='Purples')

plt.title('Nome do fornecedor por Tempo de fabricaÃ§Ã£o')
plt.xlabel('Nome do fornecedor')
plt.ylabel('Tempo de fabricaÃ§Ã£o')
```
